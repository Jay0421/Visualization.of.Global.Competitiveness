```python
data = pd.read_csv("C:/Users/hp/Desktop/2nd Semester Assignments/Data Visualization/GCI_Data.csv", delimiter=',')
data_2014 = data[data.Year == 2014]
data_2015 = data[data.Year == 2015]
data_2016 = data[data.Year == 2016]
```

```python
trace1 = go.Scatter(
    x=data_2014.GDP,
    y=data_2014['7th_pillar_Labor_market_efficiency'],
    name = "Labor Efficiency",
    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
    text= data_2014.Country
)
# second line plot
trace2 = go.Scatter(
    x=data_2014.GDP,
    y=data_2014['10th_pillar_Market_size'],
    xaxis='x2',
    yaxis='y2',
    name = "Market Size",
    marker = dict(color = 'rgba(160, 112, 20, 0.8)'),
    text= data_2014.Country
)
data = [trace1, trace2]
layout = go.Layout(
    xaxis2=dict(
        domain=[0.6, 0.95],
        anchor='y2',        
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],
        anchor='x2',
    ),
    title = 'Labor Efficiency and Market Size vs GDP'

)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
